## Analyse NBA Stats Regular Season

```python
df_stats_regular. head(5)
```

#### Corrigindo jogadores que jogam em mais de uma posição

```python
df_stats_regular['Pos'] = df_stats_regular['Pos'].apply(lambda x: x.split('-')[0] if '-' in x else x)
```

#### Filtrando os Dataframes para analisar por posição

```python

df_stats_regular_pg = df_stats_regular[df_stats_regular['Pos'] == 'PG'][['Player', 'Pos', 'PTS', 'AST','STL','TOV']]
df_stats_regular_sg = df_stats_regular[df_stats_regular['Pos'] == 'SG'][['Player', 'Pos', 'PTS', '3P%', 'STL', 'TRB']]
df_stats_regular_sf = df_stats_regular[df_stats_regular['Pos'] == 'SF'][['Player', 'Pos', 'PTS', 'TRB', 'AST', 'STL']]
df_stats_regular_pf = df_stats_regular[df_stats_regular['Pos'] == 'PF'][['Player', 'Pos', 'TRB', 'PTS', 'BLK', '2P%']]
df_stats_regular_c = df_stats_regular[df_stats_regular['Pos'] == 'C'][['Player', 'Pos', 'TRB', 'BLK', 'PTS', '2P%']]
```

#### Rankeando os jogadores

```python

df_stats_regular_pg_sorted = df_stats_regular_pg.sort_values(by=['PTS', 'AST','STL','TOV'], ascending=[False, False, False, True])
df_stats_regular_sg_sorted = df_stats_regular_sg.sort_values(by=['PTS', '3P%', 'STL', 'TRB'], ascending=[False, False, False, False])
df_stats_regular_sf_sorted = df_stats_regular_sf.sort_values(by=['PTS', 'TRB', 'AST', 'STL'], ascending=[False, False, False, False])
df_stats_regular_pf_sorted = df_stats_regular_pf.sort_values(by=['TRB', 'PTS', 'BLK', '2P%'], ascending=[False, False, False, False])
df_stats_regular_c_sorted = df_stats_regular_c.sort_values(by=['TRB', 'BLK', 'PTS', '2P%'], ascending=[False, False, False, False])
```

#### Selecionando os melhores da Regular Season

```python
df_best_regular_pg = df_stats_regular_pg_sorted.loc[df_stats_regular_pg_sorted.index[0:1], ['Player', 'Pos', 'PTS']]
df_best_regular_sg = df_stats_regular_sg_sorted.loc[df_stats_regular_sg_sorted.index[0:1], ['Player', 'Pos', 'PTS']]
df_best_regular_sf = df_stats_regular_sf_sorted.loc[df_stats_regular_sf_sorted.index[0:1], ['Player', 'Pos', 'PTS']]
df_best_regular_pf = df_stats_regular_pf_sorted.loc[df_stats_regular_pf_sorted.index[0:1], ['Player', 'Pos', 'PTS']]
df_best_regular_c = df_stats_regular_c_sorted.loc[df_stats_regular_c_sorted.index[0:1], ['Player', 'Pos', 'PTS']]
df_best_regular_players = pd.concat([df_best_regular_pg, df_best_regular_sg, df_best_regular_sf, df_best_regular_pf, df_best_regular_c])

df_best_regular_players.head(5)
```
